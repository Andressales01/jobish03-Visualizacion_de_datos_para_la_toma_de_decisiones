## Question 1
¿Is there a relation between potential and age? Does the player position affects?

```python
sns.lmplot(x = 'potential',
           y = 'age',
           data = df,
           fit_reg= True,
           hue = 'player_positions'
          )
plt.title('Relación entre potencial y edad', fontsize = 20)
```

```python
df.corr()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1) # 1er renglon, 2da columna, 1st position
sns.boxplot(data = df.loc[df['potential'] == 'weight_kg',
                          ['age']
                         ]
           )
plt.xticks(fontsize = 14, rotation =45)
plt.yticks(fontsize = 14)
ax2 = fig.add_subplot(1,2,2) # 1er renglon, 2da columna, 2nd position
sns.boxplot(data = df.loc[df['potential'] == 'heihgt_cm',
                          ['age']
                         ]
           )
plt.xticks(fontsize = 14, rotation =45)
plt.yticks(fontsize = 14)
ax1.title.set_text('weight_kg')
ax2.title.set_text('heihgt_cm')
```

```python
df['potential'].describe()
```

```python
pivota = pd.pivot_table(df,
                        values = 'potential',
                        aggfunc= 'mean',
                        columns= 'weight_kg')
```

## Conclusion
We can inffer by looking my graphics and charts that the weigth affects directly the variable of potential. We can look in the last chart that the players that weight between 70 and 90 kg are the ones with more potential.


## Question 2
Can a player´s 'weight' be determined with the
value of your 'height' and 'age'?

```python
sns.scatterplot(x = df['weight_kg'],
 y = df['height_cm'])
plt.title = ('scatter weight vs height')
```

## Observation
this can be tested by running a simple regression with 'weight' as the dependent variable and 'height' and age' as independent variables. This would allow us to check the correlation and the value of R^2 to determine if it has a good fit or not.

```python

sns.scatterplot(x = df['weight_kg'],
 y = df['age'],
 )
plt.title = ('scatter peso vs edad')
```

```python
modelA = sm.OLS.from_formula('weight_kg ~ height_cm + age', data = df)
resultA = modelA.fit()
resultA.summary()
```

## Conclusion
We can see how the graphic of height has a better regression than the one of age. We can assume that is not recommended to determine your player´s weight this way because my r^2 is 67%.
